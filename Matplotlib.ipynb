#Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x= np.linspace(0,5,11)# número de 0 até 5
x
```

```python
y= x* x
y
```

```python
plt.plot(x,y,color='r')
plt.xlabel('Eixo X')
plt.ylabel('Eixo y')
plt.title('Título')
```

```python
plt.subplots(nrows=3, ncols=3)# 3 linhas com 3 colunas
```

```python
plt.subplots(nrows=1, ncols=2)# 1 linha com 2 colunas
```

```python
plt.subplot(1,2,1)#o terceiro elemento indica com quem eu estou trabalhando, no caso vou trabalhar com o gráfico 1
plt.plot(x,y,'r--')
plt.subplot(1,2,2)#trabalhando com o segundo gráfico
plt.plot(x,y,'g*-');
```

```python
fig=plt.figure()# criando uma figura
axes=fig.add_axes([0.1,0.1,0.8,0.8])  # criando eixo

#0.1-> 10% de distancia do lado esquerdo
#0.1-> 10% de distancia da parte inferior
#0.8-> 80% largura
#0.8-> 80% altura
axes.plot(x,y)
axes.set_xlabel('Eixo X')
axes.set_ylabel('Eixo y')
```

```python
fig=plt.figure()# criando uma figura
axes1=fig.add_axes([0.1,0.1,0.8,0.8])  # criando eixo
axes2=fig.add_axes([0.2,0.3,0.3,0.3])  # criando eixo


axes1.plot(x,y)
axes1.set_xlabel('Eixo X')
axes1.set_title('Teste')
axes2.plot(x,y)
```

Subplots

```python
fig, ax= plt.subplots()# criando uma figura e um eixo simples
ax.plot(x,x**3,'b--')# utilizando o plot do eixo
ax.plot(x,x**4,'g-.')#segundo plot
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_title('Title')
```

```python
fig, ax= plt.subplots(nrows=1,ncols=2)# criando uma figura e 2 eixos
```

```python
ax #neste caso feito pelo subplots, forma uma lista de eixos
```

```python
ax[0].plot(x,y,'b--')
ax[0].set_title('Eixo 1')

ax[1].plot(x,y,'b--')
ax[1].set_title('Eixo 2')

fig
```

```python
fig,axes = plt.subplots(figsize=(8,10))
axes.plot(x,y,'r-',label='Dados 1')
axes.plot(y,x,'g-',label='Dados 2')
axes.set_title("Teste de plot")
axes.legend(loc=0)
```
